```python
import pandas as pd
import joblib
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
X_test = pd.read_csv("/content/drive/MyDrive/AeroSafeAI/X_test.csv")
y_test = pd.read_csv("/content/drive/MyDrive/AeroSafeAI/y_test.csv")
```

```python
from google.colab import files
import joblib

model = joblib.load("/content/drive/MyDrive/AeroSafeAI/models/aerosafe_model1.pkl")
print("Fin !")
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
accuracy = accuracy_score(y_test, predictions)

precision = precision_score(y_test, predictions)

recall = recall_score(y_test, predictions)

f1 = f1_score(y_test, predictions)
```

```python
print("Accuracy :", accuracy)
print("Precision :", precision)
print("Recall :", recall)
print("F1 Score :", f1)
```

```python
cm = confusion_matrix(
    y_test,
    predictions
)

print(cm)
```

```python
print(
    classification_report(
        y_test,
        predictions
    )
)
```
