```python
stations = {
    "ATL": "72219013874",
    "ORD": "72530094846",
    "DFW": "72259003927"
}

dfs = []

for airport, station in stations.items():

    url = f"https://www.ncei.noaa.gov/data/global-hourly/access/2015/{station}.csv"

    df = pd.read_csv(url)

    df["airport"] = airport

    dfs.append(df)

weather = pd.concat(dfs)

weather.head()
```

```python
weather.to_csv("weather_2015.csv", index=False)
```

```python
weather.columns
```

# Garder uniquement les colonnes utiles

```python
weather = weather[
    [
        "DATE",
        "WND",
        "VIS",
        "TMP",
        "DEW",
        "AA1"
    ]
].copy()
```

# Supression des doublons

```python
weather.drop_duplicates(inplace=True)
```

# Gérer les valeurs manquantes

```python
weather.replace(["9999", "99999", "999999"], pd.NA, inplace=True)

weather.dropna(inplace=True)
```

# Convertir la date

```python
weather["DATE"] = pd.to_datetime(weather["DATE"])

weather["YEAR"] = weather["DATE"].dt.year
weather["MONTH"] = weather["DATE"].dt.month
weather["DAY"] = weather["DATE"].dt.day
weather["HOUR"] = weather["DATE"].dt.hour
```

# Extraire la visibilité

```python
weather["VISIBILITY"] = (
    weather["VIS"]
    .str.split(",")
    .str[0]
    .astype(float)
)
```

# Extraire le vent

```python
weather["WIND_SPEED"] = (
    weather["WND"]
    .str.split(",")
    .str[3]
    .astype(float)
)
```

# Extraire la température

```python
weather["TEMPERATURE"] = (
    weather["TMP"]
    .str.split(",")
    .str[0]
    .astype(float)
    / 10
)
```

# Extraire les précipitations

```python
weather["PRECIPITATION"] = (
    weather["AA1"]
    .fillna("00,0000,1,1")
    .str.split(",")
    .str[1]
    .astype(float)
)
```

# Garder uniquement les colonnes finales

```python
weather_clean = weather[
    [
        "YEAR",
        "MONTH",
        "DAY",
        "HOUR",
        "VISIBILITY",
        "WIND_SPEED",
        "TEMPERATURE",
        "PRECIPITATION"
    ]
]
```

```python
dataset_final = pd.merge(
    flights,
    weather,
    on=["YEAR", "MONTH", "DAY", "HOUR"],
    how="left"
)
```

```python
print(dataset_final.shape)
```

```python
dataset_final.to_csv(
    "dataset_final_aerosafe.csv",
    index=False
)
```

```python
def create_risk(row):

    if row["VISIBILITY"] < 2000:
        return "Rouge"

    elif row["WIND_SPEED"] > 50:
        return "Orange"

    elif row["DEPARTURE_DELAY"] > 60:
        return "Orange"

    else:
        return "Vert"
```

```python
dataset_final["RISK_LEVEL"] = dataset_final.apply(
    create_risk,
    axis=1
)
```
