```python
import zipfile

with zipfile.ZipFile("flight-delays.zip", "r") as zip_ref:
    zip_ref.extractall("flight_data")
```

```python
df1 = pd.read_csv("flight_data/flights.csv")
df2 = pd.read_csv("flight_data/airports.csv")
df3 = pd.read_csv("flight_data/airlines.csv")
```

```python
list_drop = ['FLIGHT_NUMBER', 'TAIL_NUMBER', 'AIRLINE', 'TAXI_OUT', 'TAXI_IN', 'WHEELS_OFF', 'WHEELS_ON', 'AIR_TIME', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY', 'CANCELLED', 'CANCELLATION_REASON', 'DIVERTED']
i = 0
for i in range(0, len(list_drop)-1):
  df1.drop(list_drop[i], axis = 1, inplace = True)
```

```python
df1.head()
```

```python
df1.dropna(inplace = True)
```

```python
df1.isnull().sum()*100 / len(df1)
```

```python
# Étape A: Transformer l'entier en chaîne de 4 caractères (ex: 730 devient "0730")
df1['TIME_STR'] = df1['SCHEDULED_DEPARTURE'].astype(str).str.zfill(4)

# Étape B: Isoler les heures et les minutes
df1['HOUR'] = df1['TIME_STR'].str[:2]
df1['MINUTE'] = df1['TIME_STR'].str[2:]

# Étape C: Créer une vraie colonne Datetime combinée
df1['FLIGHT_DATETIME'] = pd.to_datetime(
    df1['YEAR'].astype(str) + '-' +
    df1['MONTH'].astype(str) + '-' +
    df1['DAY'].astype(str) + ' ' +
    df1['HOUR'] + ':' + df1['MINUTE'] + ':00'
)

# Étape D: Supprimer les colonnes temporaires pour garder le dataset propre
df1 = df1.drop(columns=['TIME_STR', 'HOUR', 'MINUTE'])
```
